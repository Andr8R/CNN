# Smoothing function sanity check

For each smoother, make sure we can reproduce the plots of the inverse CDF for various values of $\rho$ and $q$.

```python
from smoothers import *
import torch
import numpy as np

import matplotlib.pyplot as plt
```

```python
q_vals = np.linspace(0, 1, 200)
```

### Symmetric smoother

DVAE++ paper, figure 1b

```python
smoother = SymmetricSmoother(beta=5)

results = np.zeros((3, 200))

for idx_rho, rho in enumerate([0.2, 0.5, 0.8]):
    t_rho = torch.tensor(rho)
    for idx_q, q in enumerate(list(q_vals)):
        t_q = torch.tensor(q)
        results[idx_rho, idx_q] = smoother.icdf(t_rho, t_q)
        
plt.plot(q_vals, results[0,:], color='red')
plt.plot(q_vals, results[1,:], color='green')
plt.plot(q_vals, results[2,:], color='blue')
plt.xlabel("q")
plt.ylabel("Inverse CDF")
plt.ylim([0, 1])
plt.title("Spike and exponential smoother")
plt.legend(["ρ = 0.2", "ρ = 0.5", "ρ = 0.8"]);
```
