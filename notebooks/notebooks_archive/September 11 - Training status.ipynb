## VAE's training

```python
# Using the absolute path
run_ids = ["20190910_224301", "20190910_224503", "20190910_224657", "20190909_183138", "20190910_224815"]
latent_dims = [16, 32, 64, 128, 256]
```

```python
for run_id, latent_dim in zip(run_ids, latent_dims):
    dump_dir = "/home/akajal/WatChMaL/VAE/dumps/" + run_id + "/"
    training_log, val_log = dump_dir + "log_train.csv", dump_dir + "log_val.csv"
    
    print("--------------------------------------------------------------------")
    print("Plotting the training and validation log of VAE with {0} dimensions".format(latent_dim))
    print("--------------------------------------------------------------------")
    
    model_id = "ENet (" + str(latent_dim) + ")"
    # Plot training log
    plot_utils.plot_vae_training([training_log], [model_id], 
                                 {model_id:["red", "blue"]},
                                 downsample_interval=32, show_plot=True)

    # Plot validation log
    plot_utils.plot_vae_training([val_log], [model_id], 
                                 {model_id:["red", "blue"]},
                                 downsample_interval=32, show_plot=True)
```
